# Demographic analysis

If we take a look at the last columns in the dataframe, most of these are related to demographics of the voters (in the last subtask, i.e., 2.8, we made use one of these columns - CVotesU18). We also have three genre columns indicating the genres of a particular movie. We will extensively use these columns for the third and the final stage of this analysis wherein we will analyse the voters across all demographics and also see how these vary across various genres. So without further ado, let's get started with demographic analysis.

```python
# Create the dataframe df_by_genre
df_by_genre=movies.loc[:,"CVotes10":"VotesnUS"]
df_by_genre[["genre_1","genre_2","genre_3"]]=movies[["genre_1","genre_2","genre_3"]]
df_by_genre
```

```python
# Creating a column cnt and initialize it to 1
df_by_genre["cnt"]=1
```

```python
df_by_genre[["genre_1","genre_2","genre_3"]]
```

```python
# Grouping the movies by individual genres
df_by_g1=df_by_genre.groupby("genre_1").aggregate(np.sum)
df_by_g2=df_by_genre.groupby("genre_2").aggregate(np.sum)
df_by_g3=df_by_genre.groupby("genre_3").aggregate(np.sum)
```

```python
df_by_g1
```

```python
df_by_g2
```

```python
df_by_g3
```

```python
# Adding the grouped data frames and storing it in a new data frame
df_add=df_by_g1.add(df_by_g2,fill_value=0)
df_add=df_add.add(df_by_g3,fill_value=0)
df_add
```

```python
# Extracting genres with atleast 10 occurences
genre_top_10=df_add.loc[df_add["cnt"]>10]
```

```python
genre_top_10
```

```python
# Taking the mean for every column by dividing with cnt 
genre_top_10.iloc[:,0:-1]=genre_top_10.iloc[:,0:-1].divide(genre_top_10["cnt"],axis=0)
```

```python
genre_top_10
```

```python
# Rounding off the columns of Votes to two decimals
genre_top_10.loc[:,"VotesM":"VotesnUS"]=round(genre_top_10.loc[:,"VotesM":"VotesnUS"],2)
```

```python
genre_top_10
```

```python
# Converting CVotes to int type
genre_top_10[genre_top_10.loc[:,"CVotes10":"CVotesnUS"].columns]=genre_top_10[genre_top_10.loc[:,"CVotes10":"CVotesnUS"].columns].astype(int)
```

```python
genre_top_10
```

```python
# Countplot for genres
sns.barplot(x=genre_top_10["cnt"],y=genre_top_10.index)
plt.show()
```

# US vs non-US Cross Analysis

The dataset contains both the US and non-US movies. Let's analyse how both the US and the non-US voters have responded to the US and the non-US movies.

To do that , we have to Create a column IFUS in the dataframe movies. The column IFUS should contain the value "USA" if the Country of the movie is "USA". For all other countries other than the USA, IFUS should contain the value non-USA.

Now we have to draw a boxplot that shows how the number of votes from the US people i.e. CVotesUS is varying for the US and non-US movies. Making use of the column IFUS to make this plot. Similarly, making another subplot that shows how non US voters have voted for the US and non-US movies by plotting CVotesnUS for both the US and non-US movies. 

Again repeating the similar analysis but with the ratings. Making a boxplot that shows how the ratings from the US people i.e. VotesUS is varying for the US and non-US movies. Similarly, making another subplot that shows how VotesnUS is varying for the US and non-US movies. 


```python
movies['Country'].value_counts()
```

```python
# Creating IFUS column
movies['IFUS'] = movies['Country'].copy()
```

```python
movies.loc[movies['IFUS']!='USA','IFUS']='Non-USA'
movies['IFUS'].value_counts()
```

```python
# Box plot - 1: CVotesUS(y) vs IFUS(x)
# Box plot - 2 CVotesnUS(y) vs IFUS(X)
plt.figure(figsize= (20,10))
plt.subplot(1,2,1)
sns.boxplot(x='IFUS',y='CVotesUS', data = movies)
plt.subplot(1,2,2)
sns.boxplot(x='IFUS',y='CVotesnUS', data = movies)
plt.show()
```

```python
# Box plot - 3: VotesUS(y) vs IFUS(x)
#Box plot -  4: VotesnUS(y) vs IFUS(x)
plt.figure(figsize= (20,10))
plt.subplot(1,2,1)
sns.boxplot(x='IFUS',y='VotesUS', data = movies)
plt.subplot(1,2,2)
sns.boxplot(x='IFUS',y='VotesnUS', data = movies)
plt.show()
```
